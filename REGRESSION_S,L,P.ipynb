```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/tvmarketing.csv')
```

```python
df.head(5)
```

```python
X=pd.DataFrame(df['TV'])
Y=pd.DataFrame(df['Sales'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
y_pred=linreg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
Accuracy=r2_score(y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color = 'red')
plt.plot(X, linreg.predict(X), color = 'blue')
plt.title('Single Linear Regression')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

MULTIPLE LINEAR REGRESSION


```python
import pandas as pd
new=pd.read_csv('/content/drive/MyDrive/kc_house_data.csv')
```

```python
df1=new.drop('id',axis=1)
df1=df1.drop('date',axis=1)
df1=df1.drop('zipcode',axis=1)
```

```python
X=df1.drop('price',axis=1)
```

```python
Y=df1['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
lnr=LinearRegression()
```

```python
lnr.fit(x_train,y_train)
```

```python
y_pred=lnr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_pred,y_test))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red');
```

POLYNOMIAL REGRESSION

```python
import pandas as pd
p=pd.read_csv('/content/drive/MyDrive/Position_Salaries.csv')
```

```python
p.head(5)
```

```python
X=pd.DataFrame(p['Level'])
Y=p['Salary']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_poly,Y,test_size=0.2,random_state=0)
```

```python
linreg=LinearRegression()
```

```python
linreg.fit(X_train,Y_train)
```

```python
y_pred=linreg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_pred,Y_test))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color = 'red')
plt.plot(X, linreg.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Polynomial Regression,degree=4')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
