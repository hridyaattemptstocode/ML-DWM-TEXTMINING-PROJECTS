```python
import numpy as np
transaction=[]
for i in range(0, df.shape[0]):
  for j in range(0,df.shape[1]):
    transaction.append(df.values[i,j])
transaction=np.array(transaction)
transaction
```

```python
pandadf=pd.DataFrame(transaction,columns=["items"])
pandadf["incident_count"]=1
indexNames=pandadf[pandadf['items']=="nan"].index
pandadf.drop(indexNames, inplace=True)
pandadf_table=pandadf.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
pandadf_table.head(5).style.background_gradient(cmap='Greens')
```

```python
import plotly.express as px
pandadf_table["all"]="Top 50 items"
fig=px.treemap(pandadf_table.head(50),path=['all',"items"],values='incident_count', color=pandadf_table["incident_count"].head(50),hover_data=['items'],color_continuous_scale='Blues')
fig.show()
```

```python
df=pd.read_csv('/content/drive/MyDrive/Market_Basket_Optimisation.csv')
transaction=[]
for i in range(df.shape[0]):
  transaction.append([str(df.values[i,j]) for j in range(df.shape[1])])

transaction=np.array(transaction)
transaction
!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(transaction).transform(transaction)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
first30 = pandadf_table["items"].head(30).values
df=df.loc[:,first30]
df.shape
```

Implementing FPGrowth ALgorithm

```python
%pip install mlxtend --upgrade
```

```python
from mlxtend.frequent_patterns import fpgrowth
res=fpgrowth(df,min_support=0.05,use_colnames=True)
res.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules
res=association_rules(res,metric="lift",min_threshold=1)
res
```
