```python
from nltk.tokenize import word_tokenize
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("/content/drive/MyDrive/S7/Text Mining/Exp 3/BBC News.csv")
df.head()
```

```python
#Word Tokenization
data = [i for i in df.iloc[:,1]]
df_word = []
for j in range(len(data)):
    df_word.append(word_tokenize(data[j]))
```

```python
len(df_word[9])
```

```python
# Removing Stopwords
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))


#print(df_word[0:1]);
for i in range(len(df_word)):
    temp = []
    for j in df_word[i]:
        if j not in stop_words:
            temp.append(j)
    #print(temp)
    df_word[i] = temp.copy()

#print(df_word[0:1])
```

```python
len(df_word[9])
df_word[9]
```

```python
# Removing Punctuations
import string

# punctuations
punctuations=list(string.punctuation)

#print(df_word[0:1]);
for i in range(len(df_word)):
    temp = []
    for j in df_word[i]:
        if j not in punctuations:
            temp.append(j)
    #print(temp)
    df_word[i] = temp.copy()

#print(df_word[0:1])
```

```python
len(df_word[9])
df_word[9]
```

```python
# Here we achieve the Vocabulary, or wordset, (1)
wordset = []
for i in range(len(df_word)):
    wordset =  np.union1d(wordset,df_word[i])
len(wordset)
```

```python
# Function to convert each of the review to a dictionary where the keys are the words in the vocabulary,  (2)
# and the key's(word's) value is the count of that word(key) in the corresponding review
def calculateBW(wordset,doc):
  tf = dict.fromkeys(wordset,0)
  for word in doc:
      tf[word]=doc.count(word)
  return tf
```

```python
bw ={}
for i in range(len(df_word)):
    bw[i] = calculateBW(wordset,df_word[i])


#for i in range(len(bw)):

bow = pd.DataFrame(bw)
bow = bow.T

bow


#bow.to_csv('BoW.csv')
```

```python
bow.head()
```

```python
import numpy as np
senta= np.array(bow.iloc[0])
sentb= np.array(bow.iloc[1])
dot_product = np.dot(senta, sentb)

# Calculate the Euclidean norms
norm_senta = np.linalg.norm(senta)
norm_sentb = np.linalg.norm(sentb)

# Calculate the cosine similarity
cosine_similarity = dot_product / (norm_senta * norm_sentb)

print(cosine_similarity)
```

```python
similarity = []
for i in range(len(bow)):
  row=[]
  for j in range(len(bow)):
    senta = np.array(bow.iloc[i])
    sentb = np.array(bow.iloc[j])
    dot_product = np.dot(senta, sentb)

    norm_senta = np.linalg.norm(senta)
    norm_sentb = np.linalg.norm(sentb)
    cosine_similarity = dot_product / (norm_senta * norm_sentb)
    row.append(cosine_similarity)
  similarity.append(row)
```

```python
short_similarity = [row[:5] for row in similarity[:5]]
for row in short_similarity:
  formatted_row = [f"{value:.{3}f}" for value in row]
  print(" ".join(formatted_row))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

train_data = pd.read_csv('/content/drive/MyDrive/S7/Text Mining/Exp 3/BBC News Train.csv')
test_data = pd.read_csv('/content/drive/MyDrive/S7/Text Mining/Exp 3/BBC News Test.csv')

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train_data['Text'])
X_test = vectorizer.transform(test_data['Text'])
y_train = train_data['Category']
y_test = test_data['Category']

nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

nb_predictions = nb_model.predict(X_test)
print("Naïve Bayes Predicted Categories:")
print(nb_predictions)

svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

svm_predictions = svm_model.predict(X_test)
print("\nSupport Vector Machines Predicted Categories:")
print(svm_predictions)

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

knn_predictions = knn_model.predict(X_test)
print("\nK-Nearest Neighbor Predicted Categories:")
print(knn_predictions)

print("\nClassification Report for Naïve Bayes:")
print(classification_report(y_test, nb_predictions))

print("\nClassification Report for Support Vector Machines:")
print(classification_report(y_test, svm_predictions))

print("\nClassification Report for K-Nearest Neighbor:")
print(classification_report(y_test, knn_predictions))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_model_confusion_matrix(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix for ",model_name)
    print(cm)

plot_model_confusion_matrix(nb_model, X_test, y_test, 'Naïve Bayes')

plot_model_confusion_matrix(svm_model, X_test, y_test, 'Support Vector Machines')

plot_model_confusion_matrix(knn_model, X_test, y_test, 'K-Nearest Neighbor')
```
