```python
df = pd.read_csv("/content/drive/MyDrive/S7/Text Mining/Exp 2/BBC News.csv")
df.head()
```

```python
#Word Tokenization
data = [i for i in df.iloc[:,1]]
df_word = []
for j in range(len(data)):
    df_word.append(word_tokenize(data[j]))
```

```python
len(df_word[9])
```

```python
# Removing Stopwords
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))


#print(df_word[0:1]);
for i in range(len(df_word)):
    temp = []
    for j in df_word[i]:
        if j not in stop_words:
            temp.append(j)
    #print(temp)
    df_word[i] = temp.copy()

#print(df_word[0:1])
```

```python
len(df_word[9])
df_word[9]
```

```python
# Removing Punctuations
import string

# punctuations
punctuations=list(string.punctuation)

#print(df_word[0:1]);
for i in range(len(df_word)):
    temp = []
    for j in df_word[i]:
        if j not in punctuations:
            temp.append(j)
    #print(temp)
    df_word[i] = temp.copy()

#print(df_word[0:1])
```

```python
len(df_word[9])
df_word[9]
```

```python
# Here we achieve the Vocabulary, or wordset, (1)
wordset = []
for i in range(len(df_word)):
    wordset =  np.union1d(wordset,df_word[i])
len(wordset)
```

```python
# Function to convert each of the review to a dictionary where the keys are the words in the vocabulary,  (2)
# and the key's(word's) value is the count of that word(key) in the corresponding review
def calculateBW(wordset,doc):
  tf = dict.fromkeys(wordset,0)
  for word in doc:
      tf[word]=doc.count(word)
  return tf
```

```python
bw ={}
for i in range(len(df_word)):
    bw[i] = calculateBW(wordset,df_word[i])


#for i in range(len(bw)):

bow = pd.DataFrame(bw)
bow = bow.T

bow


#bow.to_csv('BoW.csv')
```

```python
#Latent Semantic Indexing(LSI)

def convert(lst):
    return ' '.join(lst)

df_new=[]
for i in range(len(df_word)):
    temp = convert(df_word[i])
    df_new += [temp]

 #print(df_new)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df_new)
(bag_of_words.todense()).shape
```

```python
bag_of_words.todense()
```

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=2)
lsa = svd.fit_transform(bag_of_words)
```

```python
import pandas as pd

topic_encoded_df = pd.DataFrame(lsa, columns = ["topic_1", "topic_2"])
topic_encoded_df["Article"] = df_new

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(topic_encoded_df[["Article", "topic_1", "topic_2"]])
```
