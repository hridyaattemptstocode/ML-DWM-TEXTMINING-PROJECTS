```python
from apyori import apriori
import numpy as np
```

```python
df.shape
```

```python
records=[]
for i in range(0,9835):
  records.append([str(df.values[i,j])for j in range(0,32)])
```

```python
association_rules=apriori(records,min_support=0.0045,min_confidence=0.55,min_lift=2,min_length=2)
association_results=list(association_rules)
association_results
```

```python
print(len(association_results))
```

```python
x=apriori(transactions = records, min_support=0.1)
p=list(x)
for i in p:
  print("Item set :",list(i[0]),"Support",i[1])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("RHS " + items[0] + " LHS ----> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("-------")
```
