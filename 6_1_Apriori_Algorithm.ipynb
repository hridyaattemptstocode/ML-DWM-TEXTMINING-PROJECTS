```python
df=pd.read_csv('/content/groceries.csv')
df
```

```python
df.shape
```

```python
records=[]
for i in range (0,9834):
             records.append([str(df.values[i,j])for j in range(0,32)])
records
```

```python
association_rules=apriori(records,min_support=0.00455,min_confidence=0.55, min_lift=2,min_length=2)
association_results=list(association_rules)
print(association_results)
```

```python
rules = apriori(records, min_support=0.0045, min_confidence=0.55, min_lift=3, min_length = 4)
results = list(rules)
results_list = []
for i in range(0, len(results)):
    results_list.append('RULE: ' + str(results[i][0]) + 'SUPPORT: ' + str(results[i][1]))
results_list
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
x=apriori(transactions = records, min_support = 0.1)
p=list(x)
for i in p :
  print("Item set :" ,list(i[0])," support:", i[1])
```
